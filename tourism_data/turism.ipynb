```python
tourist = pd.read_excel('travel_data.xls', sheet_name='number_of_arrival', header=3, index_col=0)
tourist.sample(5)
```

```python
print(tourist.shape)
tourist.info()
```

```python
# function to select rows and columns
def data_selection(data):
    df_new = pd.concat([data['Country Code'], data.loc[:,"1995":"2017"]], axis=1)
    df_new = df_new.fillna(0)
    return df_new

tourist_clean = data_selection(tourist)
tourist_clean.info()
tourist_clean.sample(5)
```

```python
# function to reshape the dataset.
def reshape_data(data):
    df = data[:]
    df.set_index(['Country Code'], inplace=True, append=True)
    df = df.stack()
    df = pd.DataFrame(df)
    df = df.reset_index()
    df = df.rename(columns={'level_2':"Year", 0:"Amount"})
    return df

tourist_reshape = reshape_data(tourist_clean)
# tourist_reshape.sample(5)
tourist_reshape.info()
tourist_reshape.shape
```

```python
# plotly tourist animation graph
map = px.choropleth(tourist_reshape,
                    locations='Country Code',
                    color='Amount',
                    # basemap_visible=False,
                    hover_name='Country Name',
                    animation_frame='Year',
                    # color_discrete_map=px.colors.sequential.Blues_r,
                    color_continuous_scale=px.colors.sequential.BuGn,
                    projection='natural earth')
map
# plotly.offline.plot(map, filename="inbound_map")
```

```python
inbound_top = px.line(tourist_reshape,
                      x = 'Year',
                      y = 'Amount',
                      color='Country Name',
                      line_group='Country Name',
                      line_shape='linear',
                      title='Top countries that attract tourist most!')

inbound_top
```
