BASIC 7 QUESTIONS ASK TO YOUR DATA.
1. HOW BIG IS THE DATA?
2. HOW DOES THE DATA LOOK LIKE?
3. WHAT IS THE DATATYPE OF COLUMNS?
4. ARE THE ANY MISSING VALUES?
5. HOW DOES THE DATA LOOK MATHEMATICALLY?
6. ARE THERE DUPLICATE VALUES?
7. HOW IS THE CORRELATION BETWEEN COLUMNS?

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/understandData/train.csv')
```

```python
# HOW BIG IS THE DATA?
df.shape
```

```python
# HOW DOES THE DATA LOOK LIKE?
df.head()
df.sample(5)  # gives random sample of your data.
```

```python
# WHAT IS THE DATA TYPE OF COLUMNS?
df.info()
```

```python
# ARE THERE ANY MISSING VALUES?
df.isna().sum()
df.isnull().sum()
```

```python
# HOW DOES THE DATA LOOK MATHEMATICALLY?
df.describe()
```

```python
# ARE THERE DUPLICATE VALUES?
df.duplicated().sum()
```

```python
# HOW IS THE CORRELATION BETWEEN COLUMNS?
df.corr()  # pearson corelation coefficient.(-1, 1)
```

$$ EDA-BIVARIATE/MULTIVARIATE $$

```python
tips = pd.read_csv('data/understandData/tips.csv')
tips.sample(5)
```

```python
titanic = pd.read_csv("data/understandData/titanic.csv")
titanic.sample(5)
```

```python
iris = pd.read_csv('data/understandData/iris.csv')
iris.sample(5)
```

```python
flights = pd.read_csv('data/understandData/flights.csv')
flights.sample(5)
```

```python
# SCATTER plot (NUMERICAL - NUMERICAL)

# bivariate
# sns.scatterplot(x='total_bill', y='tip', data=tips) 

# Multivariate.
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips) #
```

```python
# BAR PLOT (Numerical - Categorical)

# bivariate.
# sns.barplot(x='Pclass', y='Age', data=titanic)

# Multivariate
# sns.barplot(x='Pclass', y='Fare', hue='Sex', data=titanic)
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic)
```

```python
# BOX PLOT (Numerical - Categorical)

# Bivariate
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)
```

```python
# DIST PLOT (Numerical - Categorical)
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist = False)
```

```python
# HEATMAP (Categorical - Categorical)
# sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))

# calculate % of servival.
# (titanic.groupby(['Pclass', 'Sex']).mean()['Survived'] * 100).plot(kind='bar')
# titanic.groupby(['Pclass', 'Sex']).mean()['Survived'] * 100
titanic.groupby(['Pclass', 'Embarked']).mean()['Survived'] * 100
```

```python
# CLUSTER MAP.(Categorical - Categorical)
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
# PAIR PLOT (collection of scatter plot)
sns.pairplot(iris, hue='species')
```

```python
# LINE PLOT (Numerical - Numerical)
new = flights.groupby('year').sum().reset_index()
# sns.lineplot(x='year', y='passengers', data=new)

# sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))

sns.clustermap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
