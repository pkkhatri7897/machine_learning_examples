```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1").text  # gives 403 - bad requests means server reject request.
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
# GIVES VIEW PAGE INFO OUTPUT.
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
# print(soup.prettify())
```

```python
# soup.find_all('h1')[0].text
for i in soup.find_all('h2'):
    print(i.text.strip(), end=', ')
```

```python
# soup.find_all('p', class_='rating')
for i in soup.find_all('p', class_='rating'):
    print(i.text.strip(), end=", ")
```

```python
# soup.find_all('a', class_='review-count')
for i in soup.find_all('a', class_="review-count"):
    print(i.text.strip(), end=", ")
```

```python
# HERE COMES SOME ERROR BECAUSE ALL INFO ARE COMES IN ONE TAB.
for i in soup.find_all('p', class_="infoEntity"):
    print(i.text)
```

```python
company = soup.find_all("div", class_="company-content-wrapper")
name, ratings, reviews, ctype, place, years, employees = ([] for _ in range(7))

for i in company:
    name.append(i.find('h2').text.strip())
    ratings.append(i.find('p', class_='rating').text.strip())
    reviews.append(i.find('a', class_='review-count').text.strip())
    ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
    place.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
    years.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
    employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
```

```python
d = {'Name': name, 'Rating' : ratings, 'Reviews' : reviews, 'Company type' : ctype, 'Place' : place, 'Est. year' : years, 'Employees' : employees}
df = pd.DataFrame(d)
print(df.shape)
print(df)
```

```python
name, ratings, reviews, ctype, place, years, employees = ([] for _ in range(7))
final = pd.DataFrame()
for j in range(232):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(f'https://www.ambitionbox.com/list-of-companies?page={j}', headers=headers).text
    soup = BeautifulSoup(webpage, 'html.parser')
    company = soup.find_all("div", class_="company-content-wrapper")
    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)
        
        try:
            ratings.append(i.find('p', class_='rating').text.strip())
        except:
            ratings.append(np.nan)
        
        try:
            reviews.append(i.find('a', class_='review-count').text.strip())
        except:
            reviews.append(np.nan)

        try:
            ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)

        try:
            place.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
        except:
            place.append(np.nan)

        try:
            years.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
        except:
            years.append(np.nan)

        try:
            employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)

d = {'Name': name, 'Rating' : ratings, 'Reviews' : reviews, 'Company type' : ctype, 'Place' : place, 'Est. year' : years, 'Employees' : employees}
df = pd.DataFrame(d)
print(df.shape)
print(df) 
```
